```python
# Functions Warm-Up

# 1) Schreiben Sie eine Funktion, der eine Liste mit beliebig vielen Texten in Form von Stringvariablen \
#    übergeben wird. Aus dieser Liste soll ein verschachteltes Dictionary erzeugt und zurückgegeben werden, 
#    das als Keys Namen für die einzelnen Texte enthält und dessen Values Dictionaries sind, welche die Häufigkeit der Vokabularworte 
#    in den jeweiligen Texten enthalten (also das gleiche verschachtelte Dictionary, wie in der vorigen Aufgabe für 3 Texte angelegt wurde)

def text_dict(textlist):
    results = {}
    for txt in textlist[::]:
        res = txt.split()
        counts = dict()
        for word in res[::]:
            if word in counts:
                counts[word] += counts.get(word, 0) + 1
            else:
                counts[word] = 1
        name = 'text {}'.format(txt[0])
        results[name] = counts
    return results

string1 = 'the quick brown fox jumps over the lazy dog'
string2 = 'all the quick brown fox jumps over the lazy dog'
string3 = 'with the quick brown fox jumps over the lazy dog'
string4 = 'about the quick brown fox jumps over the lazy dog'

text1 = []
text1.append(string1)
text1.append(string2)
text1.append(string3)
text1.append(string4)
print(text1)

print(text_dict(text1))

# 2) Legen Sie eine Liste von Texten (Stringvariablen) an, übergeben Sie diese der oben implementierten Funktion und geben Sie das von der #    Funktion zurückgegebene verschachtelte Dictionary aus.
```
